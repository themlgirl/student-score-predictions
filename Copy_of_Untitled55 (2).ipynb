```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
#perfect relation:y=2x
def predict(x,m,b):
  return m*x+b
def cost_function(x,y,m,b):
  n=len(x)
  predictions = predict(x,m,b)
  cost=(1/n)*np.sum((predictions-y)**2)
  return cost
def gradient_descent(x,y,m,b,learning_rate,iterations):
  n=len(x)
  cost_history=[]

  for i in range(iterations):
    predictions=predict(x,m,b)
    error=predictions-y

    dm=(2/n)*np.sum(error*x)
    db=(2/n)*np.sum(error)

    m-= learning_rate*dm
    b-= learning_rate*db

    cost=cost_function(x,y,m,b)
    cost_history.append(cost)
  return m,b, cost_history
m=0
b=0
learning_rate=0.01
iterations=100
m,b,cost_history=gradient_descent(x,y,m,b,learning_rate,iterations)
print(f"traind m :{m},b:{b}")
plt.plot(range(iterations),cost_history)
plt.xlabel('iterations')
plt.ylabel('cost')
plt.title('cost function over time')
plt.show()
plt.scatter(x,y,label="data points")
plt.plot(x,predict(x,m,b),color='red',label='regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear regression')
plt.legend()
plt.show()
```
